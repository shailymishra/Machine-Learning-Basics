```python
## 100 points
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import numpy as np

########### Classifier #####################################
from sklearn.svm import SVC


X, y = make_classification(n_features=2,random_state=0,n_redundant=0)
X_train, X_test, y_train, y_test = train_test_split(  X, y, random_state=0)

############ Original Data ####################################
plt.scatter(X[:,0],X[:,1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Data')
plt.show()

clf = SVC(random_state=0)
clf.fit(X_train, y_train)
accuracy = clf.score(X_test,y_test)
print('Original Accuraacy',accuracy)


############ Data multiplied by scaler => change in decision boundary???? ####################################
X1 = 10*X
y1 = y
X_train, X_test, y_train, y_test = train_test_split(  X1, y1, random_state=0)
plt.scatter(X1[:,0],X1[:,1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Data')
plt.show()
accuracy = clf.score(X_test,y_test)
print('Original Accuraacy',accuracy)




############ Data multiplied by random scaler => change in decision boundary ####################################
randomnumber = np.random.rand(100)
randomnumber = randomnumber[:,None]
X1 = randomnumber*X
y1 = y
X_train, X_test, y_train, y_test = train_test_split(  X1, y1, random_state=0)
plt.scatter(X1[:,0],X1[:,1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Data')
plt.show()
accuracy = clf.score(X_test,y_test)
print('Original Accuraacy',accuracy)


############ Data multiplied by orthogonal matrix => change in decision boundary ####################################


from scipy.stats import ortho_group  # Requires version 0.18 of scipy
orthogonalmatrix = ortho_group.rvs(dim=100)
X1 = np.matmul(orthogonalmatrix,X)
y1 = y
X_train, X_test, y_train, y_test = train_test_split(  X1, y1, random_state=0)
plt.scatter(X1[:,0],X1[:,1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Data')
plt.show()
accuracy = clf.score(X_test,y_test)
print('Original Accuraacy',accuracy)


############ Data multiplied by rank deficient matrix => change in decision boundary ####################################

from numpy.linalg import matrix_rank
randommatrix = np.random.rand(100,100)
randommatrix[99,:] =0
print(matrix_rank(randommatrix)) # Full rank matrix
X1 = np.matmul(randommatrix,X)
y1 = y
X_train, X_test, y_train, y_test = train_test_split(  X1, y1, random_state=0)
plt.scatter(X1[:,0],X1[:,1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Data')
plt.show()
accuracy = clf.score(X_test,y_test)
print('Original Accuraacy',accuracy)
```
