```python
############# 1. PLot scatter #######################################################
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

X, y = make_classification(n_features=2,random_state=0,n_redundant=0)
print(X.shape)
plt.scatter(X[:,0],X[:,1], c=y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('matplotlib plot')
plt.show()
```
